```python
import sys
sys.path.append('../py/')
from utils import load, get_sentences
```

```python
import glob
raw_fnames = glob.glob('../data/raw/cd1/02_말뭉치/현대/구어/현대구어_말뭉치/형태분석_말뭉치/*')
len(raw_fnames)
```

```python
def summary(sentences):
    s_ = []
    for sentence in sentences:
        text, tagged = zip(*sentence)
        s_.append('%s\t%s' % (' '.join(text), ' '.join(tagged)))
    return s_
```

# 현대 구어 말뭉치

## encoding check

```python
for raw_fname in raw_fnames:
    fname = raw_fname.split('/')[-1]
    try:
        docs = load(raw_fname)
        print('-----: %s has %d lines' % (fname, len(docs)))
    except Exception as e:
        print('ERROR: %s: %s' % (fname, str(e)))
        continue
```

## Sentence format check

```python
for raw_fname in raw_fnames:
    fname = raw_fname.split('/')[-1]
    try:
        sentences = get_sentences(raw_fname)
        print('-----: %s has %d sentences' % (fname, len(sentences)))
    except Exception as e:
        print('ERROR: %s: %s' % (fname, str(e)))
        continue
```

## Raw to processed text

```python
import os
for raw_fname in raw_fnames:
    processed_fname = raw_fname.replace('../data/raw/', '../data/processed/')
    directory = '/'.join(processed_fname.split('/')[:-1])
    if not os.path.exists(directory):
        os.makedirs(directory)
    sentences = get_sentences(raw_fname)
    sentences = summary(sentences)
    with open(processed_fname, 'w', encoding='utf-8') as f:
        for sentence in sentences:
            f.write('%s\n' % sentence)
    print('done %s' % processed_fname)
```
